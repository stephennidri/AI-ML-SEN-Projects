```python
import matplotlib.pyplot as plt

# Data from Table 1
suspended_mass = [0.0048, 0.0054, 0.0076, 0.0098, 0.012]  # in kg
force = [0.047, 0.053, 0.074, 0.096, 0.12]  # Force = mass * 9.8 (N)
acceleration = [0.00023, 0.068, 0.031, 0.44, 1.06]  # in m/s^2

# Create scatter plot with linear fit
plt.figure(figsize=(8,6))
plt.scatter(force, acceleration, color='blue', label='Experimental Data', s=100)

# Linear best-fit line (F = m*a -> slope = mass of system)
import numpy as np
slope, intercept = np.polyfit(force, acceleration, 1)
fit_line = np.polyval([slope, intercept], force)
plt.plot(force, fit_line, color='red', linestyle='--', label=f'Best Fit Line (Slope = {slope:.4f} kg)')

# Plot formatting
plt.title("Force (N) vs. Acceleration (m/s)", fontsize=14)
plt.xlabel("Force (N)", fontsize=12)
plt.ylabel("Acceleration (m/s)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
